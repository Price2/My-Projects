```python
class User() : #User class with the details of each user account.
    def __init__(self, name, age, gender, ID, password):
        self.name = name
        self.age = age
        self.gender = gender
        self.ID = ID
        self.password = password
    def show_Details (self):
        print("User Details")
        print(f'user name: {self.name}, user age = {self.age}, user gender is {self.gender}, your ID is {self.ID}, your password is {self.password}')
```

```python
import re, json, time
from IPython.display import clear_output

# Made all the necessary input checks inside the bank class
class Bank(User):

    def __init__(self, name= None, age = None, gender= None,ID= None,password= None, balance = 0):    # ==== Init values given an initial value to prevent errors.
        User.__init__(self,name, age, gender, ID, password)
        self.balance = balance
        self.amount = 0
        self.password = password
        self.ID = ID


    def checkname(self): #Validates the name input to contain only letters.
        Flag = False
        try:
            while Flag == False:
                new_name = input("Please enter your name: ")
                for character in new_name:
                    # if character.isdigit():
                    if character.isalpha() == False:
                        print("Error: Make sure your name contains only letters!")
                        self.name = None
                        Flag = True
                        break
                    else:
                        continue

                if Flag == False:
                    if new_name == '':
                        continue
                    else:
                        self.name = new_name
                        return self.name
                else:
                    Flag = False

        except:
                print("Error occurred: Make sure the field is not empty and contains only letters!")
    
    def checkage(self): #Validates age input to start from 18 years old and to contain only numbers!
        while True:
            try:
            
                new_age = int(input("Please enter your age: "))
                if new_age in range (18,100) and str(new_age).isdecimal():
                    self.age = new_age
                    return self.age
                else:
                    print("Invalid number, please enter a valid age.")
                    self.age = None
                    continue
            except:
                print("Error occured: Please enter your age again!")
    
    def newpassword(self): #Registering password, must contain at-least one upper-case letter and a number
        while True:
            try:
                new_password = input('Please enter the password: ')
                if re.search('\d.*[A-Z]|[A-Z].*\d', new_password):
                    self.password = new_password
                    return self.password
                else:
                    print("Error: Password must contain at-least 1 upper case letter and 1 digit.")
            except:
                print("Error occurred: Make sure not to leave the field empty and enter a strong password.")
    
    def checkpassword(self): #Validating password before each operation to check the user's account, if password entered is invalid 3 times then return back to register/login page.
        counter = 0
        while True:
            try:
                while counter <= 3:
                    new_password = input('Please verify your password: ')
                    with open('test1.json', 'r+') as jsonfile:
                        data = json.load(jsonfile)
                    if new_password == self.password:
                        print('Verification complete.')
                        return True
                    else:
                        print("Verification Failed: Please make sure you enter your password correctly.")
                        counter +=1
                        if counter == 3:
                            return False
            except:
                print("Error occurred while entering the password, make sure the field is not empty and enter the correct password to your account.")
                counter +=1
                if counter == 3:
                    return False



    def checkgender(self): #Validates gender input to be only Male/Female and only letters to be entered
        while True:
            try:
                new_gender = input('Choose your gender "Male/Female": ')
                if (new_gender == "Male" or new_gender == "male") or (new_gender == "Female" or new_gender == "female"):
                    if new_gender.isalpha():
                        self.gender = new_gender
                        return self.gender
                else:
                    print("Incorrect input: make sure you type 'Male/Female' correctly! ")
            except:
                print("Error occured: Please make sure not to leave the field empty and to type 'Male or Female' correctly.")

    def checkID(self): #New ID during register to create a new ID and to check whether that ID has been registered before or not.
        found = False
        while True:
            try:
                new_ID = input("Please enter an ID for your account, make sure your ID is only numbers!: ")
                with open('test1.json', 'r') as jsonfile:
                    data = json.load(jsonfile)

                for i in data["users"]:
                    if new_ID == i['ID']:
                        print("Error: An account already exists with same ID.")
                        found = True
                        break
                    else:
                        found = False
                        continue

                if found == False:
                    if new_ID.isdigit() and len(new_ID) > 2:
                        self.ID = new_ID
                        return self.ID
                    else:
                        print("Incorrect input: please make sure your account ID contains only numbers and more than 2 digits.")
                else:
                    continue
            except:
                print("Error occured: something went wrong, please enter your ID again.")

    def logID(self): #ID for logging in, checks for ID entered from the user and checks if it's in the database, if it is then set the object's attributes to the user's attributes.
        count = 0
        login = False
        while True:
            try:
                login = input("Login: Please enter your ID: ")
                with open('test1.json', 'r') as jsonfile:
                    data = json.load(jsonfile)
                if count == 3:
                    return False
                for i in data["users"]:
                    if login == i['ID']:
                        password = input("Password: Please enter your password: ")
                        if password == i['password']:
                            print("Successfully logged in.\n\n")
                            self.ID = i['ID']
                            self.password = i['password']
                            self.balance = i['balance']
                            self.name = i['name']
                            self.age = i['age']
                            self.gender = i['gender']
                            return True
                        else:
                            print("Invalid password, please enter the correct password to your account.")
                    else:
                        print("Invalid ID or account does not exist.")
                        count +=1
            except:
                print("Error occured: Please make sure to enter your data correctly.")




    def main(self): #Main function, handles choices and the operations, saves data after registering, if the choice is to log-in then choose from a list the available operations.
                    #Used JSON file to create a database transforming python object to JSON string and vice versa, the database is constructed as a list of dictionaries,
                    #in which each dictionary in a list represents a user and his attributes.
                    #Tested this program slightly after finishing it, took me quite some time to finish it and fix any loopholes/bugs in it, if there's any errors/bugs,
                    #I'd appreciate the feedback.
        while True:
            try:
                print("=========Please pick an option from below:=========\n\n\n",
                "1. Register new account.\n", 
                "2. Login to your account.\n")
        
                choice = int(input())
                if choice == 1:
                    clear_output(wait=True)
                    self.checkname()
                    self.checkage()
                    self.checkgender()
                    self.newpassword()
                    self.checkID()
                    with open('test1.json', 'r+') as jsonfile:
                        data = json.load(jsonfile)
                    data["users"].append(self.__dict__)
                    with open('test1.json', 'w') as jsonfile:
                        json.dump(data, jsonfile)
                    print("Successful Registration, please login to your account.\n")
                    
                if choice == 2:
                    clear_output(wait=True)
                    if self.logID():
                        while True:
                            try:
                                print("1. Show details to your account.\n",
                                "2. Show account balance.\n",
                                "3. Withdraw amount from your account.\n",
                                "4. Deposit to your account balance.\n",
                                "5. Sign out from your account.\n",
                                "0. To exit from the program.\n")
                                select = int(input("Please choose an option from the list."))
                                if select == 1:
                                    if self.checkpassword():
                                        clear_output(wait=True)
                                        self.show_Details()
                                        time.sleep(1.5)
                                    else:
                                        clear_output(wait=True)
                                        break
                                if select == 2:
                                    if self.checkpassword():
                                        clear_output(wait=True)
                                        self.view_balance()
                                        time.sleep(1.5)
                                if select == 3:
                                    if self.checkpassword():
                                        clear_output(wait=True)
                                        self.withdraw()
                                        time.sleep(1.5)
                                if select == 4:
                                    if self.checkpassword():
                                        clear_output(wait=True)
                                        self.deposit()
                                        time.sleep(1.5)
                                        
                                if select == 5:
                                    clear_output(wait=True)
                                    time.sleep(1.5)
                                    break
                                if select == 0:
                                    clear_output(wait=True)
                                    return 'Program Successfully terminated.'
                            except:
                                print("Something went wrong please make sure to pick a valid option.")
                                time.sleep(1.5)
                    else:
                        clear_output(wait=True)
                        time.sleep(1.5)
                        continue
                    
                       
            except:
                clear_output(wait=True)
                time.sleep(1.5)
                print("Invalid input, please choose a valid option from the list.")

                

        
#======================= The operations' methods, updating the user's balance after depositing/withdrawing.

    def deposit(self):
        with open('test1.json', 'r+') as jsonfile:
            data = json.load(jsonfile)
        for i in data["users"]:
            if self.password == i["password"]:
                amount = int(input("Please enter the amount you wish to deposit: "))
                self.balance = self.balance + amount
                i['balance'] = self.balance
        with open('test1.json', 'w') as jsonfile:
            json.dump(data, jsonfile)
        print(f'Your new balance is {self.balance}')
        return self.balance

    def withdraw (self):
        with open('test1.json', 'r+') as jsonfile:
            data = json.load(jsonfile)
        for i in data["users"]:
            if self.password == i["password"]:
                amount = int(input("Please enter the amount you wish to withdraw: "))
                if self.balance > amount:
                    self.balance = self.balance - amount
                    print('New Balance is ', self.balance)
                else:
                    print('Insufficient Funds !')

    def view_balance(self):
        print(f'Your current Balance is {self.balance}')

    
        
```
