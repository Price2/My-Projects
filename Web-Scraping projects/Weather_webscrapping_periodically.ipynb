```python
from bs4 import BeautifulSoup
import requests, csv, time
import pandas as pd


#Scrap the weather temperature, humidity, visibility, air pressure,
#  wind speed, datetime data from this website
with open ('weather_data.csv', 'w', newline='', encoding='UTF-8') as weather:
    writer = csv.DictWriter(weather, fieldnames=['Weather Temperature', 'Humidity', 'visibility', 'Air Pressure', 'Wind speed', 'Time'])
    writer.writeheader()
    for _ in range(0,10):
        response = requests.get('https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt')
        soup = BeautifulSoup(response.text, 'html.parser')
        test = soup.find_all('div', attrs={'id':'current-weather'})
        weatherInfo = test[0].find('div', attrs={'class':'stats'}).get_text().replace('\r', "").replace('\n', "").replace('|', '').split()
        Temp = test[0].find('div', attrs={'class':'temp'}).get_text()
        humidity = ''.join(weatherInfo[2:3])
        visibility = ''.join(weatherInfo[4:6])
        air_pressure = ''.join(weatherInfo[7:])
        wind_speed = test[0].find('span', attrs={'class':'white'}).get_text()
        datetime = test[0].find_all('div', attrs={'class':'last-renew'})[0].get_text()
        writer.writerow({'Weather Temperature':Temp, 'Humidity': humidity, 'visibility':visibility, 'Air Pressure':air_pressure, 'Wind speed':wind_speed, 'Time':datetime})
        time.sleep(60)

```
