# Most pop zeft 5ararh

```python
most_Global_Sales = df.sort_values('Global_Sales',ascending=False)[:10]

def name_and_year(row):
    return row['Name'] + " (" + str(int(row['Year'])) + ") " 

most_Global_Sales['Name and Year'] = most_Global_Sales.apply(name_and_year,axis = 1)

plt.pie(most_Global_Sales['Global_Sales'],labels = most_Global_Sales['Name and Year'],autopct='%.1f%%')
plt.show()
```

```python
#all games
most_Global_Sales = df.sort_values('Global_Sales',ascending=False)#[:1000]

Genre = most_Global_Sales['Genre'].value_counts()

plt.pie(Genre,labels = Genre.index,autopct='%.1f%%')
plt.show()
```

```python
#top 100
most_Global_Sales = df.sort_values('Global_Sales',ascending=False)[:100]

Genre = most_Global_Sales['Genre'].value_counts()

plt.pie(Genre,labels = Genre.index,autopct='%.1f%%')
plt.show()
```
