#### <b>Getting first 20 games with most Global_Sales, then applying feature engineering to extract new information from the existing columns.

<li><b>Since the first video game developed, which studios are the most dominant in video games?</li>

```python
most_Global_Sales = df.sort_values('Global_Sales',ascending=False)[:20]

def name_and_year(row):
    return row['Name'] + " (" + str(int(row['Year'])) + ") " 

most_Global_Sales['Name and Year'] = most_Global_Sales.apply(name_and_year,axis = 1)

plt.pie(most_Global_Sales['Global_Sales'],labels = most_Global_Sales['Name and Year'],autopct='%.1f%%')
plt.show()
```

#### <b>Apply feature engineering to put the games in different categories than what is already there, where all games selling the most will get an AAA rate followed by A and B ratings and C

#### <b>Fixing the Years from float to int values

```python

df['Year'] = df['Year'].apply(lambda row: int(row))
df
```

```python
def add_ratings(row):
    if row['Global_Sales'] >= 0.480000:
        return 'AAA'
    if  (row['Global_Sales'] <=0.170000) & (row['Global_Sales'] >= 0.060000):
        return 'B'
    return 'C'
     
    # print(row[ row['Global_Sales']>=30] )

df['Rating'] = df.apply(add_ratings,axis=1)
df
```

```python
df['Rating'].value_counts()
```

<li><b>Over the course of years, what were the most successful genres in terms of sales?</li>

```python
genre_in_years =df.pivot_table(index='Year', columns='Genre', values='Global_Sales')
genre_in_years
```

```python
fig,ax = plt.subplots(1,1,figsize=(25,20))
my_colors=[(0.2,0.3,0.3),(0.4,0.5,0.4),(0.1,0.7,0),(0.1,0.7,0)]

sns.heatmap(genre_in_years, cmap=my_colors ,linewidth=0.1, linecolor=(0.1,0.2,0.2), ax=ax, annot=True, fmt='.2g')

colorbar = ax.collections[0].colorbar
M=genre_in_years.max().max()
colorbar.set_ticks([1/8*M,3/8*M,6/8*M])
colorbar.set_ticklabels(['C','B','AAA'])

plt.show()
```

<font color='orange' size=3><b>Doing a pivot table to show a relation of what are the most successful genres over the years and their revenus.</b></font>

<font color='orange' size=3><b>Remark: The most noticeable years which includes most selling games were from 1983 to 1990</b></font>

<font color='orange' size=3><b>Remark: The following genres were the most selling from 1983-1990, Shooter, Platform, Puzzle</b></font>

<font color='orange' size=3><b>Remark: There is a decline in sales over the course of years, sales declining from 1990 till 2020 and started stablizing from 1990 </b></font>
